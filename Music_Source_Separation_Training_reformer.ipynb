```python
#@markdown # Install
from google.colab import output
import subprocess, threading, time, os, textwrap

def spinner(stop_event):
    frames = "|/-\\"
    i = 0
    while not stop_event.is_set():
        print(f"\rInstalling... {frames[i % len(frames)]}", end="")
        time.sleep(0.1)
        i += 1
    print("\r", end="")

def install(commands, cwd=None):
    stop = threading.Event()
    t = threading.Thread(target=spinner, args=(stop,))
    t.start()

    try:
        for cmd in commands:
            subprocess.check_output(
                cmd,
                shell=True,
                cwd=cwd,
                stderr=subprocess.STDOUT,
            )

        stop.set()
        t.join()
        output.clear()
        print("✅ Install successful")

    except subprocess.CalledProcessError as e:
        stop.set()
        t.join()
        print("\n❌ Install failed\n")
        print(f"Command: {cmd}\n")
        print(e.output.decode())
        raise


# --------------------------------------------------
# Prepare requirements text
# --------------------------------------------------
req_text = textwrap.dedent("""
mutagen==1.47.0
ml_collections==1.1.0
numpy>=1.26.0
pandas==2.2.2
scipy
loralib
tqdm
segmentation_models_pytorch==0.3.3
timm
audiomentations
pedalboard
omegaconf
beartype
rotary_embedding_torch==0.3.5
einops
demucs
torchmetrics==0.11.4
spafe==0.3.2
protobuf
torch_audiomentations
asteroid==0.7.0
auraloss
torchseg
PoPE-pytorch>=0.0.15
""")

# Ensure working directory
os.makedirs("/content", exist_ok=True)

# --------------------------------------------------
# Run multi-step install
# --------------------------------------------------
install(
    [
        # 1️⃣ Clone repo
        "git clone -b colab-inference-2026 https://github.com/jarredou/Music-Source-Separation-Training",
        # 2️⃣ Write requirements.txt inside cloned folder
        f"echo '{req_text}' > Music-Source-Separation-Training/requirements.txt",
        # 3️⃣ Make ckpts folder
        "mkdir -p Music-Source-Separation-Training/ckpts",
        # 4️⃣ Install Python dependencies
        "pip install -r Music-Source-Separation-Training/requirements.txt",
    ],
    cwd="/content",
)
%cd '/content/Music-Source-Separation-Training/'
!wget 'https://raw.githubusercontent.com/jarredou/Music-Source-Separation-Training-Colab-Inference/refs/heads/main/models.json' -q
```

```python
import os
import json
import torch
import yaml


# ─────────────────────────────────────────────
# YAML helpers
# ─────────────────────────────────────────────

class IndentDumper(yaml.Dumper):
    def increase_indent(self, flow=False, indentless=False):
        return super().increase_indent(flow, False)


def _tuple_constructor(loader, node):
    return tuple(loader.construct_sequence(node))

yaml.SafeLoader.add_constructor(
    'tag:yaml.org,2002:python/tuple', _tuple_constructor
)


# ─────────────────────────────────────────────
# Download
# ─────────────────────────────────────────────

def download_file(url: str, save_as: str | None = None, dest_dir: str = 'ckpts') -> str:
    os.makedirs(dest_dir, exist_ok=True)

    filename  = save_as if save_as else os.path.basename(url.split('?')[0])
    file_path = os.path.join(dest_dir, filename)

    if os.path.exists(file_path):
        print(f"[skip] '{filename}' already exists.")
        return file_path

    print(f"[download] {filename} …")
    try:
        torch.hub.download_url_to_file(url, file_path)
        print(f"[done]  '{filename}'")
    except Exception as exc:
        print(f"[error] download failed: {exc}")
        raise

    return file_path


# ─────────────────────────────────────────────
# Config editor  (works for every model_type)
# ─────────────────────────────────────────────

_OVERLAP_KEYS = {'num_overlap', 'overlap'}
_ROFORMER_TYPES = {'bs_roformer', 'mel_band_roformer'}


def _set_nested(data: dict, target_keys: set, value) -> bool:
    changed = False
    for k, v in data.items():
        if k in target_keys:
            data[k] = value
            changed = True
        elif isinstance(v, dict):
            changed |= _set_nested(v, target_keys, value)
    return changed


def conf_edit(config_path: str, chunk_size: int, overlap: int, patch_chunk_size: bool = True) -> None:
    with open(config_path, 'r') as fh:
        data = yaml.load(fh, Loader=yaml.SafeLoader)

    # ── ensure use_amp exists ──
    if 'use_amp' not in data.get('training', {}):
        data.setdefault('training', {})['use_amp'] = True

    # ── chunk_size (Roformer models only) ──
    if patch_chunk_size:
        data.setdefault('audio', {})['chunk_size'] = chunk_size

    # ── batch_size ──
    data.setdefault('inference', {})['batch_size'] = 1

    # ── overlap (model-type agnostic) ──
    if not _set_nested(data, _OVERLAP_KEYS, overlap):
        data.setdefault('inference', {})['num_overlap'] = overlap

    print(f"Config used: chunk_size={chunk_size if patch_chunk_size else 'N/A'}, overlap={overlap}, batch_size=1, BigShifts={BigShifts}")

    with open(config_path, 'w') as fh:
        yaml.dump(data, fh,
                  default_flow_style=False,
                  sort_keys=False,
                  Dumper=IndentDumper,
                  allow_unicode=True)


# ─────────────────────────────────────────────
# Model loader  (reads models.json)
# ─────────────────────────────────────────────

def load_model(
    model_name:  str,
    chunk_size:  int,
    overlap:     int,
    registry:    str = 'models.json',
    dest_dir:    str = 'ckpts',
) -> tuple[str, str, str]:
    with open(registry, 'r') as fh:
        models: dict = json.load(fh)

    if model_name not in models:
        available = '\n  '.join(models.keys())
        raise ValueError(f"Unknown model '{model_name}'.\nAvailable:\n  {available}")

    meta = models[model_name]
    model_type = meta['model_type']

    config_path = download_file(
        meta['config']['url'],
        save_as=meta['config']['save_as'],
        dest_dir=dest_dir,
    )
    ckpt_path = download_file(
        meta['checkpoint']['url'],
        save_as=meta['checkpoint']['save_as'],
        dest_dir=dest_dir,
    )

    patch_chunk_size = model_type in _ROFORMER_TYPES
    conf_edit(config_path, chunk_size, overlap, patch_chunk_size=patch_chunk_size)

    return model_type, config_path, ckpt_path


# ════════════════════════════════════════════════════════════════
#  Cell 1 – Setup
# ════════════════════════════════════════════════════════════════
%cd '/content/Music-Source-Separation-Training/'


# ════════════════════════════════════════════════════════════════
#  Cell 2 – Parameters  (edit these)
# ════════════════════════════════════════════════════════════════

# ── I/O ──────────────────────────────────────────────────────────
input_folder  = '/content/drive/MyDrive/INPUT'  #@param {type:"string"}
output_folder = '/content/drive/MyDrive/OUTPUT'               #@param {type:"string"}

# ── Model ────────────────────────────────────────────────────────
import json

with open('models.json', 'r') as fh:
    _models = json.load(fh)

model_list = list(_models.keys())

model = model_dropdown.value

# ── Separation options ────────────────────────────────────────────
extract_instrumental = True   #@param {type:"boolean"}
export_format        = 'wav FLOAT'  #@param ['wav FLOAT', 'flac PCM_16', 'flac PCM_24']
BigShifts            = 1      #@param {type:"slider", min:1, max:11, step:1}
overlap              = 13      #@param {type:"slider", min:2, max:40, step:1}
use_tta              = True  #@param {type:"boolean"}

# ── Roformer tuning ───────────────────────────────────────────────
chunk_size = "587412"  #@param [88200, 112455, 132300, 156555, 176400, 352800, 485100, 529200, 588800, 587412, 661500, 749259] {allow-input: true}

# ── Output naming ─────────────────────────────────────────────────
filename_template_preset = 'subdir per song:  MySong/vocals.flac  ← default' #@param ['subdir per song:  MySong/vocals.flac  ← default', 'subdir per stem:  vocals/MySong.flac', 'flat — underscore: MySong_vocals.flac']
_PRESETS = {
    'subdir per song:  MySong/vocals.flac  ← default': '{file_name}/{instr}',
    'subdir per stem:  vocals/MySong.flac':             '{instr}/{file_name}',
    'flat — underscore: MySong_vocals.flac':            '{file_name}_{instr}',
}
filename_template = _PRESETS[filename_template_preset]
print(f"filename_template = '{filename_template}'")


# ════════════════════════════════════════════════════════════════
#  Cell 3 – Resolve model & run inference  (no edits needed)
# ════════════════════════════════════════════════════════════════

# Derive CLI flags from export_format
flac_file = export_format.startswith('flac')
pcm_type  = export_format.split(' ')[1] if flac_file else None

# Download files + patch config
model_type, config_path, ckpt_path = load_model(
    model_name  = model,
    chunk_size  = int(chunk_size),
    overlap     = int(overlap),
    registry    = 'models.json',
    dest_dir    = 'ckpts',
)

# Build optional CLI flags
flags = ' '.join(filter(None, [
    '--extract_instrumental' if extract_instrumental else '',
    '--flac_file'            if flac_file            else '',
    '--use_tta'              if use_tta               else '',
    f'--pcm_type {pcm_type}' if pcm_type             else '',
    f'--filename_template {filename_template!r}',
]))

# Run
!python inference.py \
    --model_type         {model_type} \
    --config_path        '{config_path}' \
    --start_check_point  '{ckpt_path}' \
    --input_folder       '{input_folder}' \
    --store_dir          '{output_folder}' \
    --bigshifts          {int(BigShifts)} \
    {flags}
```

**INST-Mel-Roformers like v1/1x/2** have switched output file names - files labelled as vocals are instrumentals <br>
But e.g. not v1e - so if you uncheck extract_instrumentals for it, only one stem called "other" will be rendered, and it will be instrumental.<br><br>
**Mel Karaoke by becruily and Duality** models output 2 stems, and don't need "**extract_instrumental**" option enabled (inverted stem will rather have worse quality, than the model output, plus you won't end up with 3 output files).
<br><br>
**TTA** - results in longer separation time, "it gives a little better SDR score but hard to tell if it's really audible in most cases". <br> it “means "test time augmentation", (...) it will do 3 passes on the audio file instead of 1. 1 pass with be with original audio. 1 will be with inverted stereo (L becomes R, R become L). 1 will be with phase inverted and then results are averaged for final output. ” - jarredou
<br><br>
**Overlap** - higher means longer separation time. 4 is already balanced value, 2 is fast and some people still won't notice any difference. Normally there's not point going over 8.<br><br>
**Chunk_size** - most models use the default 485100 (besides Beta6X - 529200, [BS-Roformer SW 6 stems](https://colab.research.google.com/github/jarredou/Music-Source-Separation-Training-Colab-Inference/blob/main/Colab_Inference_BSRofo_SW_fp16.ipynb) - 588800 [882000 will probably crash on free T4], and Amane's 4 stems Large - 661500), and achieves the highest SDR with this value (higher than training chunk), but some people occasionally use also 112455 to get different results.<br><br>
